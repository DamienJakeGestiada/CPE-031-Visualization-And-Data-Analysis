```python
df = pd.read_excel('/content/SampleData.xlsx')
df
```

```python
df3 = df[['Region','Item','Unit Cost']]

df3 = df3[df3['Unit Cost'] > 10]
df3
```

```python
import seaborn as sns

heat_data = df3.pivot_table(
    index="Region",
    columns="Item",
    values="Unit Cost",
    aggfunc="sum"
).fillna(0)

# Plot heatmap
plt.figure(figsize=(12, 8))
sns.heatmap(heat_data, cmap="Reds", linewidths=0.5)

plt.title("Heat Map of Unit Cost by Region and Item", fontsize=16, pad=20)
plt.xlabel("Item", fontsize=12)
plt.ylabel("Region", fontsize=12)
plt.show()
```

Interpretation:

The heat map shows how unit costs vary significantly between items and regions.  Higher costs are shown by darker red cells, and lower costs are represented by lighter colors.  In all, some products are constantly more expensive per unit throughout several regions, while others continue to be reasonably priced.

```python
df4 = df[['Units','OrderDate']]

df4 = df4[df4['Units'] > 10]
df4
```

```python
# Group by City and sum counts
city_trend = df4.groupby('OrderDate')['Units'].sum()

# Plot line chart
plt.figure(figsize=(12, 6))
plt.plot(city_trend.index, city_trend.values, marker='o', linestyle='-', color='blue')

# Add labels and title
plt.title('Total Units Ordered by Date')
plt.xlabel('OrderDate')
plt.ylabel('Units amount')

# Rotate x-axis labels for readability
plt.xticks(rotation=0)

# Show chart
plt.show()
```

Interpretation:

The trend of the total units ordered over time is displayed in the line chart.  The quantity of units generally shows an increasing trend, with unusual shifts.The times of July and February saw the largest and lowest orders, respectively.  This raises the possibility of demand-related trends.
